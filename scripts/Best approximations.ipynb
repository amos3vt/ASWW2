```python
from bert_serving.client import BertClient
bc = BertClient()
```

```python
embeddings = []
for i in [i for i in expert_tags.name]:
    embeddings.append(bc.encode([i]))
    print('ran')
```

```python
big_embeddings = []
zooniverse = [i.split('"')[1] for i in zooniverse_tags.Names]
for i in zooniverse:
    big_embeddings.append(bc.encode([i]))
```

```python
def cosine(vec1, vec2):
    return np.dot(vec1, vec2) / \
            (np.linalg.norm(vec1) * np.linalg.norm(vec2))
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarities = []
corresponding = []

for i in range(0, len(big_embeddings)):
    index = 0
    maxi = 0
    for j in range(0, len(embeddings)):
        temp = cosine_similarity(np.array(big_embeddings[i]), embeddings[j])
        if maxi < temp:
            maxi = temp
            current = expert_tags.name[j]
    similarities.append(maxi)
    corresponding.append(current)
            
        
    #maxi = np.max([cosine_similarity(np.array(big_embeddings[k]), embeddings[i]) for i in range(0, len(embeddings))])
    #similarities.append(maxi)
```

```python
final = pd.DataFrame({'Zooniverse tag': zooniverse, 'Expert correspondance': corresponding, 'similarity': similarities}).sort_values(by = 'similarity', ascending = False)
```
