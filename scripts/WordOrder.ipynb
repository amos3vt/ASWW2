```python
import matplotlib.pyplot as plt
import itertools
import numpy as np
from bert_serving.client import BertClient
bc = BertClient()
from transformers import BertTokenizer
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
```

```python
# Get all permutations of words in a sentence
# Compare the embedding for each of these permutations with the
#   embedding for the original sentence

perm = list(itertools.permutations(["change", "the", "ordering", "of", "words."]))

orig_s = " ".join(perm[0])
orig_embed = bc.encode([orig_s])[0]

cosine_similarity = []

for i in range(1,len(perm)):
    s = " ".join(perm[i])
    embed = bc.encode([s])[0]
    cs = np.dot(orig_embed, embed)/(np.linalg.norm(orig_embed)*np.linalg.norm(embed))
    cosine_similarity.append(cs)
    print(cs)
```

```python
# For reference, compare also compare the embedding for the
#   orignal sentence with the embedding of a completely
#   different sentence.

diff_s = "What you're referring to as Linux, is in fact, GNU/Linux, or as I've recently taken to calling it, GNU plus Linux."
diff_embed = bc.encode([diff_s])[0]
print(np.dot(orig_embed, diff_embed)/(np.linalg.norm(orig_embed)*np.linalg.norm(diff_embed)))
```
